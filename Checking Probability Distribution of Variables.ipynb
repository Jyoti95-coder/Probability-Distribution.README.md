**For a very good ML model, we always have to follow one basic and the most important assumption:**
    
**The variables that are going into the ML model,must follow normal distribution.**

**Initially all the variables follow some random distributions, we have to bring them near to normal distribution using some   transformations.


**Different types of transformations are used:**

**1.Logarithmic Transformation (preferred)**

**2. Exponential transformation**

**3. Square transformations**

**4. Square root transformations**

**5. Cube root transformation**

**6. Cube root transformation**
